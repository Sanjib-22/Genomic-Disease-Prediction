```python
import matplotlib.pyplot as plt
```

```python
labels = [
    "Model",
    "Regularization Type",
    "Regularization Strength (C)",
    "Solver",
    "Class Weight",
    "Max Iterations",
    "Cross-Validation",
    "Scoring Metric",
    "Feature Selection"
]

values = [
    "Logistic Regression",
    "L-1 (LASSO)",
    "10.0",
    "saga",
    "balanced",
    "10000",
    "5-Fold Stratified",
    "ROC_AUC",
    "SelectKBest"
]

fig, ax = plt.subplots(figsize=(8, len(labels)))
ax.axis('off')


table = plt.table(
    cellText=[[l, v] for l, v in zip(labels, values)],
    colLabels=["Component", "Setting"],
    loc='center',
    cellLoc='left',
    colWidths=[0.4, 0.6]
)

table.auto_set_font_size(False)
table.set_fontsize(9.5)
table.scale(1, 1.4)

ax.set_title("Hyper-parameter Settings of the 10:90 Dataset Model", y=0.7, fontsize=10, pad=20)
plt.tight_layout()
plt.show()
```
