MODEL TRAINING AND TESTING

```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.linear_model import LogisticRegressionCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report, ConfusionMatrixDisplay, roc_auc_score, roc_curve, confusion_matrix 
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
import random
import os
```

```python
# File path

result_dir = os.path.join('..', 'Outputs', 'T2D_results')
target_dir = os.path.join('..', 'Final Datasets', 'T2D_final')
raw_file_dir = os.path.join('..', 'Raw_Datasets', 'T2D_data')

# Final File path 
snp_file = os.path.join(raw_file_dir, 'T2D_top2000_cleaned.csv')
pheno_final_extended = os.path.join(target_dir, 'T2D_aligned_phenotypes_final_extended.csv')
pheno_final_path = os.path.join(target_dir, 'T2D_aligned_phenotypes_final.csv') 
geno_final_path = os.path.join(target_dir, 'T2D_aligned_genotypes_final.csv')
logreg_model = os.path.join(result_dir, 'T2D_selectKbest_logreg_model.joblib')
logreg_coeff = os.path.join(result_dir, 'logreg_selectKbest_coefficients.csv')
logreg_model_AS = os.path.join(result_dir, 'T2D_selectKbest_logreg_model_AgeSex.joblib')
logreg_coeff_AS = os.path.join(result_dir, 'logreg_selectKbest_coefficients_AgeSex.csv')
model_perf = os.path.join(result_dir, 'T2D_selectKbest_model_performance.png')
model_perf_AS = os.path.join(result_dir, 'T2D_selectKbest_model_performance_AgeSex.png')
cf_matrix = os.path.join(result_dir, 'T2D_confusion_matrix.png')
prs_csv = os.path.join(result_dir, 'T2D_prs_scores.csv')
prs_png = os.path.join(result_dir, 'T2D_PRS_distribution_fixed.png')
```

```python
# Loading the data
X = pd.read_csv(geno_final_path, index_col=0)
y = pd.read_csv(pheno_final_path).set_index("sample_id")["label"]

#Aligning the samples
common = X.index.intersection(y.index)
X = X.loc[common]
y = y.loc[common]

gwas = pd.read_csv(snp_file)
gwas_filtered = gwas[gwas["p_value"]<1e-5]
gwas_filtered["chr_pos"] = gwas_filtered["chrom"].astype(str) + "_" + gwas_filtered["pos"].astype(str)
gwas_snps = gwas_filtered["chr_pos"]

X_gwas = X.loc[:, X.columns.intersection(gwas_snps)]
print(f"\nSNPs after gwas p-value filtering : {X_gwas.shape[1]}")

# Selecting top SNPs using ANOVA F-test
selector = SelectKBest(score_func=f_classif, k=min(78, X_gwas.shape[1]))
X_selected = selector.fit_transform(X_gwas, y)

# Retaining column names
selected_snps = X_gwas.columns[selector.get_support()]
X_selected_df = pd.DataFrame(X_selected, index=X_gwas.index, columns=selected_snps)
```

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_selected_df, y, test_size=0.10, stratify=y, random_state=42
)
```

```python
# Train Logistic Regression with L1 penalty
print("Training LogisticRegressionCV...")
logreg = LogisticRegressionCV(
    Cs=10,
    cv=10,
    penalty='l1',
    solver='saga',
    scoring='roc_auc',
    max_iter=10000,
    n_jobs=2,
    random_state=42
).fit(X_train, y_train)
```

```python
# Prediction and evaluation
y_pred_prob_train = logreg.predict_proba(X_train)[:, 1]
y_pred_prob = logreg.predict_proba(X_test)[:, 1]
train_auc = roc_auc_score(y_train, y_pred_prob_train)
test_auc = roc_auc_score(y_test, y_pred_prob)
print(f"Train AUC: {train_auc:.3f}")
print(f"Test AUC: {test_auc:.3f}")

y_pred_binary = (y_pred_prob > 0.5).astype(int)
accuracy = accuracy_score(y_test, y_pred_binary)
print(f"\nAccuracy (threshold 0.5): {accuracy:.3f}")

# Step 3: Optional detailed report
print("\nClassification Report:\n")
print(classification_report(y_test, y_pred_binary))

# Step 7: Save model and coefficients
joblib.dump(logreg, logreg_model)
pd.Series(logreg.coef_.flatten(), index=X_selected_df.columns, name="coefficients")\
  .to_csv(logreg_coeff)

# Step 8: Visualize ROC and Risk Score Distribution
fpr, tpr, _ = roc_curve(y_test, y_pred_prob)

plt.figure(figsize=(10, 4))

# ROC Curve
plt.subplot(1, 2, 1)
plt.plot(fpr, tpr, label=f"AUC = {test_auc:.3f}")
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Type-2 Diabetes ROC Curve (SNP only)")
plt.legend()

# Risk Score Histogram
plt.subplot(1, 2, 2)
plt.hist(y_pred_prob[y_test == 0], bins=30, alpha=1, label="Controls", color="#065a83", edgecolor='black', linewidth=0.5, zorder=3)
plt.hist(y_pred_prob[y_test == 1], bins=30, alpha=1, label="RA Cases", color="#d75b25", edgecolor='black', linewidth=0.5, zorder=2)
plt.grid(axis='y', alpha=0.3, linestyle='--', zorder=1)
plt.xlabel("Predicted Probability")
plt.ylabel("Count")
plt.title("Risk Score Distribution for Rheumatoid Arthritis (SNP Only)")
plt.legend()
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.tight_layout()
plt.savefig(model_perf, dpi=500, bbox_inches='tight')
plt.show()

# Step 9: Confusion Matrix
print("\nConfusion Matrix (threshold = 0.5):")
print(confusion_matrix(y_test, y_pred_prob > 0.5))
```

```python
#Taking only Age And Sex

# 1. Load phenotypes (has age, sex, label)
df = pd.read_csv(pheno_final_extended).set_index("sample_id")

# 2. Define age+sex as features, label as target
df["sex"] = df["sex"].map({"M":0, "F":1})
X_age_sex = df[["age", "sex"]]
y = df["label"]

# 3. Split
X_train, X_test, y_train, y_test = train_test_split(
    X_age_sex, y, stratify=y, test_size=0.10, random_state=42
)

# 4. Train and Evaluate
model = LogisticRegressionCV(Cs=10, cv=10, penalty='l1', solver='saga', max_iter=10000, n_jobs=2, random_state=42)
model.fit(X_train, y_train)

auc = roc_auc_score(y_test, model.predict_proba(X_test)[:, 1])
print(f"Age + Sex Only AUC: {auc:.3f}")
```

```python
#Taking sex and age as covariates

X = pd.read_csv(geno_final_path, index_col=0)
phenos = pd.read_csv(pheno_final_extended).set_index("sample_id")

common = X.index.intersection(phenos.index)
X = X.loc[common]
phenos = phenos.loc[common]

phenos["sex_bin"] = phenos["sex"].map({"M":0, "F":1})

y = phenos["label"]
age = phenos["age"]
sex = phenos["sex_bin"]
```

```python
gwas = pd.read_csv(snp_file)
gwas["chr_pos"] = gwas["chrom"].astype(str) + "_" + gwas["pos"].astype(str)
gwas_snps = gwas[gwas["p_value"]<1e-5]["chr_pos"]
X_gwas = X.loc[:, X.columns.intersection(gwas_snps)]
print(f"GWAS-filtered SNPs: {X_gwas.shape[1]}")

selector = SelectKBest(score_func=f_classif, k=min(78, X_gwas.shape[1]))
X_selected = selector.fit_transform(X_gwas, y)
selected_snps = X_gwas.columns[selector.get_support()]
X_selected_df = pd.DataFrame(X_selected, index=X_gwas.index, columns=selected_snps)

X_selected_df["age"] = age
X_selected_df["sex"] = sex
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_selected_df, y, test_size=0.10, stratify=y, random_state=42
)
```

```python
print("Training LogisticRegressionCV...")
logreg = LogisticRegressionCV(
    Cs=10,
    cv=10,
    penalty='l1',
    solver='saga',
    scoring='roc_auc',
    max_iter=10000,
    n_jobs=2,
    random_state=42
).fit(X_train, y_train)
```

```python
# Prediction and evaluation
y_pred_prob_train = logreg.predict_proba(X_train)[:, 1]
y_pred_prob = logreg.predict_proba(X_test)[:, 1]
train_auc = roc_auc_score(y_train, y_pred_prob_train)
test_auc = roc_auc_score(y_test, y_pred_prob)
print(f"Train AUC: {train_auc:.3f}")
print(f"Test AUC: {test_auc:.3f}")

y_pred_binary = (y_pred_prob > 0.5).astype(int)
accuracy = accuracy_score(y_test, y_pred_binary)
print(f"\nAccuracy (threshold 0.5): {accuracy:.3f}")

# Step 3: Optional detailed report
print("\nClassification Report:\n")
print(classification_report(y_test, y_pred_binary))

# Step 7: Save model and coefficients
joblib.dump(logreg, logreg_model_AS)
pd.Series(logreg.coef_.flatten(), index=X_selected_df.columns, name="coefficients")\
  .to_csv(logreg_coeff_AS)

# Step 8: Visualize ROC and Risk Score Distribution
fpr, tpr, _ = roc_curve(y_test, y_pred_prob)

plt.figure(figsize=(10, 4))

# ROC Curve
plt.subplot(1, 2, 1)
plt.plot(fpr, tpr, label=f"AUC = {test_auc:.3f}")
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Type-2 Diabetes ROC Curve (SNP+Age+Sex)")
plt.legend()

# Risk Score Histogram
plt.subplot(1, 2, 2)
plt.hist(y_pred_prob[y_test == 0], bins=30, alpha=1, label="Controls", color="#065a83", edgecolor='black', linewidth=0.5, zorder=3)
plt.hist(y_pred_prob[y_test == 1], bins=30, alpha=1, label="Cases", color="#d75b25", edgecolor='black', linewidth=0.5, zorder=2)
plt.grid(axis='y', alpha=0.3, linestyle='--', zorder=1)
plt.xlabel("Predicted Probability")
plt.ylabel("Count")
plt.title("Risk Score Distribution for Type-2 Diabetes (Age+Sex+SNP)")
plt.legend()
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.tight_layout()
plt.savefig(model_perf_AS, dpi=500, bbox_inches='tight')
plt.show()

# Step 9: Confusion Matrix
print("\nConfusion Matrix (threshold = 0.5):")
print(confusion_matrix(y_test, y_pred_prob > 0.5))
```
